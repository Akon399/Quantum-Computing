```python
# Notes & Lecture by Akhona Njeje.
# Date : 20 July 2023.
# Topic : Single Qubit, Quantum Gates.
# Constructing a quantum state with 1 Qbit.



from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
q = QuantumRegister(1) # Initialising the number of Qbits.
c = ClassicalRegister(1) # 1 classical bits.
qc = QuantumCircuit(q,c)
```

```python
qc.draw('mpl') # matplotlib = mpl form. 
```

```python
from qiskit.quantum_info import Statevector  
psi1 = Statevector(qc)
psi1.draw('latex')
```

```python
# Lets visualise the state using the Q SPHERE.

from qiskit.visualization import plot_state_qsphere
plot_state_qsphere(qc)   # The relative state of the face is zero, the probability is 100% of measuring zero.
```

```python
from qiskit.quantum_info import DensityMatrix
rho1 = DensityMatrix(qc)
rho1.draw('latex',prefix='\\rho_1 = ')
```

```python
# Lets visualise the density matrix.

from qiskit.visualization import plot_state_city
plot_state_city(qc)
```

```python
from qiskit.visualization import plot_state_hinton
plot_state_hinton(qc)
```

```python
from qiskit.visualization import plot_state_paulivec
plot_state_paulivec(qc,title='Initial Density Matrix as linear combination of Pauli gates')
```

```python
from qiskit.visualization import plot_bloch_multivector
plot_bloch_multivector(qc)
```

```python
purity = rho1.purity()
print(purity)
```
